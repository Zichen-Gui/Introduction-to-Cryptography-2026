## Introduction
In the class, you saw SHA256 as an example of a collision resistant hash function. There are, in fact, many more desgins of collision resistant hash functions before SHA256. But unfortunately, many of them are broken (and hence, the popularity of SHA256 as the "next best thing").
In this programming assignment, you will break two historical hash functions called MD4 and MD5 (MD: Message-Digest). MD5 in particular, is still very popular today despite the attacks on it.
